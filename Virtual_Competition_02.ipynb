```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
i = 1
kf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)

# Loop through each fold of the cross-validation
for train_index, test_index in kf.split(X, y):
    print('Fold {} of {} in Stratified K-Fold'.format(i, kf.n_splits))

    # Split the data into training and validation sets for this fold
    xtr, xvl = X.iloc[train_index], X.iloc[test_index]
    ytr, yvl = y.iloc[train_index], y.iloc[test_index]

    # Create a logistic regression model
    model = LogisticRegression(random_state=1)

    # Train the model on the training data
    model.fit(xtr, ytr)

    # Make predictions on the validation data
    pred_test = model.predict(xvl)

    # Calculate and print the accuracy score for this fold
    score = accuracy_score(yvl, pred_test)
    print('Accuracy Score:', score)

    # Increment the fold counter
    i += 1

# After the loop, you can use the trained model to make predictions on the test set
pred_test = model.predict(test_dataset)

# Also, you can obtain predicted probabilities for the last validation set
pred_prob = model.predict_proba(xvl)[:, 1]
```
